```python
# Importing all of my libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Making the dataframes
diabetes = pd.read_csv('diabetes_data.csv')
stroke = pd.read_csv('stroke_data.csv')
hypertension = pd.read_csv('hypertension_data.csv')
```

```python
# Snapshot of diabetes dataframe
diabetes.head(5)
```

Diabetes column definitions:

Age: Age of patient; 1=18-24, 9=60-64

Sex: Sex of patient; 1: male, 0: female

HighChol: 0=no high cholesterol, 1=high cholesterol

CholCheck: 0=no cholesterol check in 5 years, 1=yes cholesterol check in 5 years

BMI: Body Mass Index

Smoker: Has the patient smoked at least 100 cigarettes in your entire life? 0=no, 1=yes

HeartDiseaseorAttack: coronary heart disease (CHD) or myocardial infarction; 0=no, 1=yes

PhysActivity: Physical activity in past 30 days (excluding job); 0=no, 1=yes

Fruits: consume fruit 1 or more times per day; 0=no, 1=yes

Veggies: Consume vegetables 1 or more times per day; 0=no, 1=yes

HvyAlcoholConsump: adult men >=14 drinks per week, adult women >=7 drinks per week; 0=no, 1=yes

GenHlth: General health scaled 1=5; 1=excellent, 2=very good, 3=good, 4= fair, 5=poor

MentHlth: days of poor mental health scale 1-30 days

PhysHlth: physical illness or injury days in past 30 days; scale 1-30

DiffWalk: Difficulty walking or climbing stairs?; 0=no, 1=yes

Stroke: 0=no, 1=yes

HighBP: 0=no high BP, 1=High BP

Diabetes: 0=no diabetes, 1=diabetes

```python
# Snapshot of stroke dataframe
stroke.head(5)
```

Stroke column definitons:

sex: Patient's gender; 1=male, 0=female
age: Patient's age in years
hypertension: patient has ever had hypertension (1) or not (0)
heart_disease: patient has ever had heart_disease(1) or not (0)

```python
# Snapshot of hypertension dataframe
hypertension.head(5)
```

Hypertension column definitions:

age: patient's age in years
sex: patient's gender; 1=male, 0=female
cp: Chest pain type; 0=asymptomatic, 1=typical angina, 2=non-anginal pain
trestbps: Resting blood pressure (mm Hg)
chol: Serum cholesterol in mg/dl
fbs: if the patient's fasting, blood sugar > 120 mg/dl; 1=yes, 0=no

target: Whether the patient has hypertension; 1=yes, 0=no

Now that you have everything setup, put any code that you use to get to know the dataframe and its rows and columns better in the cell below. You can use whatever techniques you like, except for visualizations. You will put those in a separate section.

```python
# Getting data types for the dataframes
print(diabetes.info(), stroke.info(), hypertension.info())

# Diabetes are all floats
# Stroke is a mix of floats and ints
# Hypertension is also a mix of floats and ints
```

```python
# Looking at unique values in each column to see if I can get rid of any that are all the same
print(f"Diabetes: \n{diabetes.nunique()}\n Stroke: \n{stroke.nunique()}\n Hypertension: \n{hypertension.nunique()}")

# Looks like all the information is mostly unique for all dataframes. 
# No column in any dataframe as a value of 1, and the ones that have a 2 are yes/no.
```

```python
# Looking for null values in each column for each dataframe
print(diabetes.isnull().sum())
print(stroke.isnull().sum())
print(hypertension.isnull().sum())

# The diabetes dataframe is clean of null values
# The stroke dataframe has null values in the sex column, probably should remove since there is plenty of other complete data
# The hypertension dataframe has null values in the sex column, which I will remove since there is other complete data
```

```python
# Drop null observations from stroke and hypertension
stroke = stroke.dropna()
hypertension = hypertension.dropna()
```

```python
# Look for duplicate data in each dataframe
diabetes_dup = diabetes[diabetes.duplicated()]
stroke_dup = stroke[stroke.duplicated()]
hyp_dup = hypertension[hypertension.duplicated()]

print("Number of duplicates in diabetes: ", diabetes_dup.shape)
print("Numer of duplicates in stroke: ", stroke_dup.shape)
print("Number of duplicates in hypertension: ", hyp_dup.shape)

# There are no duplicates in stroke or hypertension, but there appear to be 6672 duplicates in diabetes.
```

```python
# Dropping the duplicate values in diabetes
print(diabetes.shape)
diabetes= diabetes.drop_duplicates()
diabetes.shape
```

```python
plt.figure(figsize=(20,10))
c=diabetes.corr()
sb.heatmap(c,cmap="BrBG",annot=True)
c
```
